<center><h1>Preprocessing</h1></center>

Podaci koje je prikupio univerzitet Johns Hopkins odnose se na period od izbijanja pandemije novog koronavirusa (Covid-19) od januara 2020. do 10. marta 2023., kada je obustavljeno dalje prikupljanje. Ali, podaci sa brojevima oporavljenih još su ranije obustavljeni, 4. avgusta 2021., tako da smo analizu ograničili na period januar 2020. - avgust 2021. Podaci su za sve države svijeta, kao i posebno za SAD.

Podaci su skinuti sa adrese: https://github.com/CSSEGISandData/COVID-19

```python
# Importovanje neophodnih modula
# Pandas - za sređivanje podataka (data wrangling)
import pandas as pd
```

```python
# Fajlovi su dati u raznim formama. Mi koristimo najpogodnije, u formi vremenskih serija.
# Oni se u arhivu nalaze na lokaciji ./COVID-19-master/csse_covid_19_data/csse_covid_19_time_series
# Otvaramo fajlove sa podacima. Kada ih otvorimo, oni će biti u formatu pandas DataFrame.
confirmed = pd.read_csv('./data/JHU/time_series_covid19_confirmed_global.csv')
deaths = pd.read_csv('./data/JHU/time_series_covid19_deaths_global.csv')
recovered = pd.read_csv('./data/JHU/time_series_covid19_recovered_global.csv')
```

```python
# Osnovni podaci o redovima (broj) i kolonama (indeks, naziv, broj validnih unosa (koji nisu nedostajuće vrijednosti) i vrsta/tip podataka).
confirmed.info()
# Prikaz prvih nekoliko redova tabele (pošto nismo uneli vrijednost u zagradu, biće ih 5)
confirmed.head()

# Isto radimo i za preostale tabele
deaths.info()
deaths.head()

recovered.info()
recovered.head()
```

```python
# Izbacujemo suvišne kolone
confirmed.drop(['Province/State', 'Lat', 'Long'], axis=1, inplace=True)
deaths.drop(['Province/State', 'Lat', 'Long'], axis=1, inplace=True)
recovered.drop(['Province/State', 'Lat', 'Long'], axis=1, inplace=True)
```

```python
# Provjeravamo ima li nedostajućih vrijednosti - ako ih ima, rezultat će biti veći od 0.
sum(confirmed.isnull().sum(axis=1).to_list())
sum(deaths.isnull().sum(axis=1).to_list())
sum(recovered.isnull().sum(axis=1).to_list())
```

```python
# Sabiramo vrijednosti za svaki dan ponaosob, kako bismo dobili brojeve slučajeva za svjetsku populaciju.
confirmed_world = confirmed.drop('Country/Region', axis=1).sum()
deaths_world = deaths.drop('Country/Region', axis=1).sum()
recovered_world = recovered.drop('Country/Region', axis=1).sum()

confirmed_world
deaths_world
recovered_world
```

```python
# U tabeli recovered vidjeli smo da su podaci u posljednjim danima jednaki 0. To vjerovatno znači da su
# ranije prekinuli da ih sakupljaju. Provjeravamo otkada nema unosa.
recovered_world.index[recovered_world > 0][-1]

# Prethodne dve serije su kumulativne, pa da proverimo da li je i ova treća
recovered_world.loc[recovered_world > 0]
```

```python
# Trebaju nam i uobičajene serije sa zabilježenim dnevnim vrijednostima.
confirmed_daily = confirmed_world.diff().fillna(confirmed_world)
confirmed_daily

deaths_daily = deaths_world.diff().fillna(deaths_world)
deaths_daily

recovered_daily = recovered_world[:'8/4/21'].diff().fillna(recovered_world)
recovered_daily
```

```python
# Provjeravamo ispravnost - suma svih vrijednosti treba biti jednaka poslednjoj kumulativnoj sumi iz prošle serije
# osim u slučaju oporavljenih, gdje je reper 4. avgust 2021.
confirmed_daily.sum() == confirmed_world[-1]
deaths_daily.sum() == deaths_world[-1]
recovered_daily.sum() == recovered_world.loc[recovered_world > 0][-1]
```

```python
# Da bismo mogli da izvedemo sve tražene pokazatelje, skratićemo prve dvije serije na dužinu treće
# Dakle, period za koji ćemo imati sve potrebne podatke biće 22. januar 2020. - 4. avgust 2021.
confirmed_daily = confirmed_daily[:561]
deaths_daily = deaths_daily[:561]

confirmed_daily
deaths_daily
```

```python
# Trebaće nam i kumulativne serije za proračune, pa ih ubacujemo u skraćenom obliku, do 4. avgusta 2021.
confirmed_cumulative = confirmed_world[:561]
recovered_cumulative = recovered_world[:561]
deaths_cumulative = deaths_world[:561]

confirmed_cumulative
recovered_cumulative
deaths_cumulative
```

```python
# Spajamo serije u jedan DataFrame, gdje ćemo dodavati dalje kolone
data = pd.concat({'confirmed_daily': confirmed_daily, 
                    'confirmed_cumulative': confirmed_cumulative, 
                    'deaths_daily': deaths_daily, 
                    'deaths_cumulative': deaths_cumulative, 
                    'recovered_daily': recovered_daily,
                    'recovered_cumulative': recovered_cumulative},
                   axis=1)
data
```

```python
# Dakle, imamo broj registrovanih, broj oporavljenih i broj umrlih - dnevno i kumulativno.
# Sada ćemo napraviti serije koje nam nedostaju.

# Ubacujemo kolonu sa prevalencom. Računamo je tako što od kumulativnih vrijednosti registrovanih slučajeva
# oduzmemo kumulativne vrijednosti oporavljenih i umrlih
data['prevalence_daily'] = data['confirmed_cumulative'] - data['deaths_cumulative'] - data['recovered_cumulative']

# Ubacujemo kolonu sa incidencom. Računamo je na dnevnom nivou, sa faktorom skaliranja od 100.000, tako da će 
# dobijena vrijednost biti izražena u jedinici "broj osoba na 100.000 stanovnika dnevno".
# Pošto nema standardizovanih procjena broja svjetskog stanovništva u formi dnevnih vremenskih serija, uzimamo
# aritmetičku sredinu godišnjeg prosjeka za 2020. i 2021. 
# (po podacima sa: https://www.worldometers.info/world-population/#table-historical)
# Dobijene vrijednosti zaokružujemo na dvije decimale.
scaling_factor = 100000
population = (7840952880 + 7909295151) // 2
data['incidence_rate'] = ((data['confirmed_daily'] / population) * scaling_factor).round(2)

# Ubacujemo kolonu sa mortalitetom. Računamo je slično kao i incidencu.
data['mortality'] = ((data['deaths_daily'] / population) * scaling_factor).round(2)

# Ubacujemo i kolone sa vrijednostima CFR (Case Fatality Rate) - dnevnim i kumulativnim.
data['cfr_daily'] = ((data['deaths_daily'] / data['confirmed_daily']) * 100).round(2)
data['cfr_cumulative'] = ((data['deaths_cumulative'] / data['confirmed_cumulative']) * 100).round(2)

# Pogledajmo šta smo dobili
data.info()
data.head()
```

```python
# Grafički prikaz
data.plot(subplots=True, layout=(6, 2), figsize=(12, 20), sharex=False)
```

```python
# Na grafiku gdje imamo oporavljene na dnevnom nivou, primjećuje se velika negativna vrijednost.
# To se dešava zato što neke epidemiološke službe katkada vrše korekcije (revizije), a to rade
# odjednom za duži period. Ali, to može značajno uticati na preciznost prognoze.
# Ovo ćemo popraviti tako što ćemo negativnu vrijednost zamijeniti prosjekom prethodnog i narednog dana.

# Najprije pronalazimo koji je to dan:
data[data['recovered_daily'] < 0].dropna()
```

```python
# Zatim mijenjamo vrijednost onako kako smo naveli
data['recovered_daily']['12/14/20'] = (data['recovered_daily']['12/13/20'] + data['recovered_daily']['12/15/20']) // 2
```

```python
# Da vidimo kako sada izgleda grafik
data['recovered_daily'].plot(figsize=(12, 6))
```
